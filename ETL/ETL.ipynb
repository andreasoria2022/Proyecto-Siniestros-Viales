Vamos a proceder a realizar un proyecto referido a Siniestro Viales. Para lo cual vamos a crear un dataset desde los archivos Excel

Procederemos a importar las librerias neceseria 

```python
import pandas as pd
```

Vamos a crear un dataset df_homicidios a partir del archivo homicidios.xlsx

```python
df_homicidios = pd.read_excel('../Data/homicidios.xlsx')
```

Vamos a ver el dataset

```python
df_homicidios
```

Voy a ver la información del dataset df_homicidios

```python
df_homicidios.info()
```

Detectamos que el dataset contiene 696 filas y 21 columnas. Y luego de observar la informacion detectamos que no posee nulos 

Vamos a ver si el dataset contiene duplicados

```python
duplicados = df_homicidios.duplicated()
print(duplicados)
```

No hay duplicados en el dataset df_homicidios

Eliminaré las columnas AAAA, MM y DD porque la información que contienen estas columnas está contenida en la columna FECHA

```python
# Eliminar la columna 'AAAA'
df_homicidios = df_homicidios.drop('AAAA', axis=1)
```

```python
# Eliminar la columna 'AAAA'
df_homicidios = df_homicidios.drop('MM', axis=1)
```

```python
# Eliminar la columna 'AAAA'
df_homicidios = df_homicidios.drop('DD', axis=1)
```

```python
df_homicidios
```

Voy a cambiar el tipo de dato de la Columna COMUNA de entero a objet. 

```python
# Cambiar el tipo de la columna 'COMUNA' a 'object'
df_homicidios['COMUNA'] = df_homicidios['COMUNA'].astype('object')
```

```python
df_homicidios.info()
```

Reemplazaré el tipo de calle en donde dice GRAL PAZ por AVENIDA

```python
df_homicidios['TIPO_DE_CALLE']= df_homicidios['TIPO_DE_CALLE'].replace({'GRAL PAZ': 'AVENIDA'})
```

```python
df_homicidios
```

Vamos a crear un csv a partir del dataset df_homicilios

```python
df_homicidios.info()
```

```python
df_homicidios.to_csv('homicidios.csv')
```

Creare un dataset df_lesiones a partir del archivo lesiones.xlsx

Voy a ver la informacion de los datos del dataset df_lesiones

Vemos que el dataset no tiene nulos en su columna 'id'. Y que el dataset df_lesiones tiene 23785 filas y 27 columnas

Procederé a ver si posee duplicados teniendo en cuenta la columna id

```python
duplicados = df_lesiones.duplicated(subset='id')
print(duplicados)
```

No hay duplicados en el dataset df_lesiones

Elimino las columnas 'aaaa'(año),'mm' (mes )y 'dd (día) porque este dato se encuentra en la columna fecha

```python
# Eliminar la columna 'aaaa'
df_lesiones = df_lesiones.drop('aaaa', axis=1)
```

```python
# Eliminar la columna 'mm'
df_lesiones = df_lesiones.drop('mm', axis=1)
```

```python
# Eliminar la columna 'dd'
df_lesiones = df_lesiones.drop('dd', axis=1)
```
